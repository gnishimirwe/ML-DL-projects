```python
import nltk
nltk.download('punkt')
```

```python
path='/content/drive/MyDrive/1661-0.txt'
text= open(path).read().lower()
len(text)
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer('\s+', gaps = True)
```

```python
text_token= tokenizer.tokenize(text)
```

```python
len(text_token)
```

```python
import re
corpus=[]
for i in range (1,len(text_token)):
  one_word= re.sub('[^a-zA-Z]', ' ', text_token[i])
  corpus.append(one_word)
```

```python
from keras_preprocessing.text import Tokenizer
k_token= Tokenizer()
k_token.fit_on_texts(corpus)
encoded_data= k_token.texts_to_sequences(corpus)[:]
len(encoded_data)
```
